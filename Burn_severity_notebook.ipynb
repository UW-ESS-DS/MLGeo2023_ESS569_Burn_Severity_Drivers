# Bootleg Fire: Evaluating the Role of Fuel Treatments in Mitigating Fire Severity near Sycan Marsh, Oregon


### SCIENTIFIC MOTIVATION
### Study's importance
Despite the extensive literature documenting the mitigating effects of fuel treatments on burn severity, the complex and dynamic nature of wildfires presents significant interpretive challenges. Identifying site-specific, key drivers of burn severity and the role of fuel treatments in mitigating burn severity are crucial tasks. Such insights guide the strategic allocation of limited management resources and lay the groundwork for future research. 

### Background 
#### Fire suppression and fuel accumulation
The Bootleg fire was ignited by lightning in July 2021 and burned a total of 413,766 acres over 41 days in the Fremont-Winema National Forest, Southcentral Oregon (NWCC). The wildfire exhibited extreme fire behavior. As reported in the New York Times on July 19, 2021, from an interview with Marcus Kauffman, a spokesperson for the Oregon Department of Forestry, “The fire is so large and generating so much energy and extreme heat that it is changing the weather, […] normally the weather predicts what the fire will do. In this case, the fire is predicting what the weather will do” ([Fountain 2021](https://www.nytimes.com/2021/07/19/climate/bootleg-wildfire-weather.html)).  The fire spread through forested areas that were treated with prescribed fire (Rx), mechanical thinning (Tx), a combination of these two fuel treatments (TxRx), and areas that were not treated at all. Fuel treatments have been conducted repeatedly during the past ~ 15 years through a collaboration between TNC, USFS, and the Klamath Tribes, offering the opportunity to study their effectiveness across diverse prescriptions and timing of implementation.

Wildfire’s extreme behavior in fuel-limited ecosystems is generally the result of a history of fire suppression, logging, gold mining, grazing, and the exclusion of Indigenous fire by Euro-American settlements since 1850. These factors have significantly altered the historic fire regimes of dry-forest ecosystems across the western United States (Fry and Stephens 2006; Wallace Covington 2000; Cova et al. 2023; Haugo et al. 2019). Continuous accumulation of biomass resulting from fire suppression combined with a rapidly warming climate has led to uncharacteristic high-severity fires (Hanan et al. 2021; Halofsky, Peterson, and Harvey 2020; Wallace Covington 2000; Allen et al. 2002; van Mantgem et al. 2013), and unusual stand-replacing fires have been increasing in terms of frequency and extent in dry forest (Westerling et al. 2006; Miller et al. 2009).
Paradoxically, by becoming more effective at controlling and extinguishing wildfire we have created a condition of fire deficit (Haugo et al. 2019) and intensified the wildfire problem (Agee and Skinner 2005). Haugo et al. (2019) reported that over the last three decades, only one-tenth of the area which historically would have been expected to burn in the forests of Washington and Oregon actually did. Letting wildfires burn, where possible, and increasing prescribed fire treatments are two ways to address the fire deficit   (Vaillant and Reinhardt 2017). 


#### Fuel treatments in the context of other environmental conditions

In fuel-limited ecosystems, where fuel has accumulated due to anthropogenic disturbances, fuel treatments are employed to reduce fuel loads and to restore previous fire regimes, promoting forest resilience. 
Factors determining fuel treatment effectiveness include vegetation dynamics (productivity, composition, structure, distribution), disturbance regimes (abiotic and biotic), suppression location and type, fire weather, location and source of fire ignition, fuel characteristics (type, amount, continuity, conditions, availability), topography, and climate. Thus, fuel treatment importance (i.e. effectiveness in mitigating burn severity) must be evaluated in the context of other environmental covariates (Prichard et al. 2020; Cansler et al. 2022; Cansler and McKenzie 2014; McKinney et al. 2022) and their continuous dynamic interactions with the landscape (Peterson 2002; Johnstone et al. 2016).


#### Burn severity drivers 

Major drivers of burn severity fall under five main categories: fuel (e.g. forest structure, vegetation type, etc.) topography, weather, climate, and management. 
The choice of specific predictor variables within each of the five categories depends on data availability, study site location and extent, time of the fire, and research question/s. Burn severity can be treated as a continuous or as categorical variable (e.g., low, moderate, high), and machine learning (ML) ensemble methods are employed to investigate the importance of its drivers at a global and local scale. Methods include random forests classification/regression, and boosted regression trees. These models are then applied to compare burn severity under different environmental conditions (Cansler et al. 2022), to investigate the heterogeneity of the local importance of each predictor (Povak et al. 2020), or to inform about the importance of each driver in the context of specific environmental settings at the time of the fire (Prichard et al. 2020; S. A. Parks et al. 2018). Studies have reported model performance with a coefficient of determination (R-squared) ranging from 50 to 65%. Generally, these analyses are conducted using the R programming language. 

#### Remote sensing methods used to estimate burn severity

Fire severity can be analyzed across various spatial scales by employing fire severity metrics calculated using remote sensing technologies. Presently, the predominant remote sensing techniques used to assess fire severity and establish correlations with fuel treatment efficacy entail computing burn severity indices derived from moderate/fine-resolution satellite imagery that can range from 30 to 10 meters depending on the satellite and the index used in the analysis. An important aspect to consider when interpreting satellite-derived burn severity values is their limited ability to capture a wide range of biophysical fire effects. This occurs because the biophysical complexity existing on the ground is reduced to a single value originating from the differencing of pre- and post-fire average spectral responses assigned to each pixel. For example, numerous small dead trees could produce the same reflectance value as a single, large dead tree (Furniss et al. 2020), despite their vastly distinct impact on ecosystem functions and carbon cycling (Lutz et al. 2018). Additionally, large tree canopies may block the passive detection of smaller, understory trees, constraining even further the reliability of satellite-derived fire severity metrics. Given these limitations, satellite-derived burn severity thresholds are informed by field surveys right after fire. In the context of forest ecology, burn severity is quantified in the field using the composite burn index (CBI; Key and Benson 2006). This burn-severity metric ranges from 0 to 3, with CBI = 0 reflecting no change due to fire and CBI = 3 reflecting the highest degree of fire-induced ecological change (S. Parks et al. 2018). For this study, we will be using the relative burn ratio (RBR) (S. Parks, Dillon, and Miller 2014) derived from Sentinel-2 Level 2A data (processed to Bottom of Atmosphere) (Howe et al. 2022) calculated with Google Earth Engine at a pseudo 10m resolution (NIR band = 10 m; SWIR2 band = 20 m but resampled at 10m). 


#### Overarching goal and objectives 

The purpose of this study is to conduct site-level analyses across a relatively small geographic area (matching the pre-fire LiDAR collection), which burned in the Bootleg fire to investigate key drivers of burn severity and the role of fuel treatments in mitigating burn severity. The overarching goal will be achieved by:
1) *Evaluate the global and local importance of burn severity drivers.*
2) Evaluate local interactions of most important burn severity drivers, and especially between prescribed fire and other covariates.
3) Evaluate the relationship between spatial progression of fire, and treatment types along with landscape features (e.g. riparian and wetland areas, topography), and firefighting operations.

Given the scope of this project, our analysis will be limited to the first objective and the interaction between prescribed fire and the most interactive covariate.

### Machine Learning
[Random forests regression](https://scikit-learn.org/stable/modules/generated/sklearn.ensemble.RandomForestRegressor.html) with [Recursive Feature Elimination](https://scikit-learn.org/stable/modules/generated/sklearn.feature_selection.RFECV.html) using the [scikit-learn package](https://scikit-learn.org/stable/index.html).

Random Forests (RF) regression was the chosen machine learning method, which is a supervised machine learning method. Random Forests analysis was done by following some of the methods illustrated in chapters 2.6 and 3.6 in the [MLGeo book](https://geo-smart.github.io/mlgeo-book/about_this_book/about_this_book.html) and other online resources, including documentation explaining packages and functions used in this analysis. The main ML package utilized was Scikit-learn, specifically the sklearn.ensemble module and the RandomForestRegressor class.  

First, we double-checked for any missing values (NaNs), which there were none, then we defined predictor features (X) and target variable (y).  
The dataset was split in 80% train and 20% test, and stratified by burn severity thresholds (Parks et al. (2018), Table 7) applicable to forested landscapes in the western US. Thresholds represented low, moderate, and high severity classes to assure an equal proportion of each burn severity class. 

The next step was to choose hyperparameters that would be used to generate our model. Key hyperparameters in the RandonForestRegressor class include:
n_estimators: number of (decision) trees in the forest,

max_depth: max depth of each tree,

min_samples_leaf: minimum number of samples required to be at a leaf node,

max_features: number of features to consider at each node split,

bootstrap: (T/F) use bootstrapped samples (true) or entire dataset (false). Bootstrapping makes the model less sensitive to the training data by creating random subsets with replacement.

We set up lists of possible options for each of the key hyperparameters (except bootstrap, which was set to True) and used the GridSearchCV class to find the best set of parameters for the model’s performance. In the first run, max_depth and max_features hyperparameters included “None”, which was chosen by the algorithm over the other options. To test the effect of other max_depth and max_features values, we ran a second iteration without “None”, and compared results to decided which model to keep. 

To cut down the number of variables being fed into the RF, we used Recursive feature elimination with cross-validation (RFECV; 10-fold) to select the most important features (trimmed to 9 variables from 24). Finally, the RF model was created, fit to the selected train data, and evaluated on train and test data.

The performance was evaluated using R-squared and RMSE functions built into sklearn.metrics. 
Comparison between R-squared and RMSE derived from train and test were used to evaluate if the model under/over-fit the data.

This model was not expected to generalize well out of the designated study area, since it was trained with spatially specific variables and under unique environmental conditions. Additionally, this is an informative model built with the purpose of finding key drivers of burn severity and understanding the role of fuel treatments in mitigating burn severity. This model won’t be used to make predictions.

### Evaluating features's importance
Since the objective of this study is to identify the key drivers of burn severity and evaluate the impact of fuel treatments relative to other environmental covariates (while accounting for possible interactions), choosing the right method to quantify variables' importance on a global and local scale is a crucial step.

Permutation or Gini importance are a global measure evaluating the importance of a feature based on its contribution to the overall model performance. While this is useful, a better understanding of variables' importance and interactions can be achieved by conducting SHapley Additive exPlanation (SHAP) analysis.This can result in a different ordering of feature importance compared to permutation/Gini methods if features interact with one another in complex ways.

"The goal of SHAP is to explain the prediction of an instance x by computing the contribution of each feature to the prediction. The SHAP explanation method computes Shapley values from coalitional game theory. The feature values of a data instance act as players in a coalition. Shapley values tell us how to fairly distribute the “payout” (= the prediction) among the features. A player can be an individual feature value, e.g. for tabular data. A player can also be a group of feature values. For example to explain an image, pixels can be grouped to superpixels and the prediction distributed among them. One innovation that SHAP brings to the table is that the Shapley value explanation is represented as an additive feature attribution method, a linear model." ([Interpretable Machine Learning](https://christophm.github.io/interpretable-ml-book/shap.html)). 

The [SHAP package](https://shap.readthedocs.io/en/latest/index.html) is used to explain a wide range of ML models. We will use it to interpret the local and global (aggregated) importance of each predictor variabale that was incuded in our final RF regression model, and to explore which variable more strongly interacted with prescribed fire (Rx). This intercation can be easily visualized using the built-in plotting function ```shap.plots.scatter``` as we will show later in this notebook. 

To put it succinctly, within the context of this study, SHAP values are useful for: 
- providing explanations for the prediction of an individual observation by computing the contribution of each feature to the prediction.
- evaluating both local interpretability (importance of features for individual predictions) and global interpretability (aggregated importance across all predictions).
- considering interaction effects between predictors.

Calculating key distribution parameters and assessing normality using Shapiro-Wilk test. 

````python
# Descriptive statistics 
analog_vars = data.drop(columns=['Unnamed: 0', 'RBR', 'X', 'Y', 'Rx', 'EVT_PP']) # isolating continuous variables for analysis
analog_vars_2 = data.drop(columns=['Unnamed: 0', 'X', 'Y', 'Rx', 'EVT_PP']) # keeping RBR for descriptivr stats and histograms
stats = {} #instantiate dict

for var in analog_vars_2:
    mean = np.mean(data[var]) #calculate mean
    median = np.median(data[var]) # calculate median
    variance = scipy.stats.variation(data[var]) # calculate variance
    kurtosis = scipy.stats.kurtosis(data[var]) # calculate kurtosis
    skewness = scipy.stats.skew(data[var])  # calculate skewness
    ShapiroWilk = scipy.stats.shapiro(data[var])[1] # calculate normality via Shapiro-Wilk
    stats[var] = {"mean": mean,
                "median": median,
                "variance": variance,
                "kurtosis": kurtosis,
                "skewness": skewness,
                "Shapiro-Wilk": ShapiroWilk} # enter var stats into dict

stats_df = pd.DataFrame.from_dict(stats)
display(stats_df.T) # show transpose of dataframe for easier readability

# According to Shapiro-Wilk p-values, none of the variables are normally distributed
````

````python
# Prepare data for train and test splits 

# Define predictor features (X) and target variable (y)
X = df.drop(columns=['RBR'])  # 'RBR' is the target variable
y = df['RBR']

# Burn severity classes for stratification 
thresholds = [135, 301] # low, medium, high burn severity 
labels = np.digitize(y, thresholds, right=True)

# Check labels 
label_counts = np.bincount(labels)

# Print the counts for each label
for label, count in enumerate(label_counts):
    print(f"Label {label}: {count} occurrences")

# Label 0: 1401 occurrences
# Label 1: 274 occurrences
# Label 2: 150 occurrences

# Split data in train and test by burn severity classes (labels)
X_train, X_test, y_train, y_test = train_test_split(X, labels, test_size=0.2, random_state=42, stratify=labels)
````

Running RF regression analysis including "None" in max_depth and max_features hyperparameters. 

````python
# Random forests regeression model with feature elimination 

# Step 1: Hyperparameter Tuning (Number of Trees, Maximum Depth, Minimum Samples per Leaf, and Number of Predictors)
param_grid = {
    'n_estimators': [200, 300, 500],  # Number of trees in the forest
    'max_depth': [10, 20, 30, 40, 'None'],  # Maximum depth of each tree
    'min_samples_leaf': [5, 10, 15, 20],  # Minimum samples per leaf node
    'max_features': ['sqrt', 'log2', 'None'],  # Number of predictors to consider at each split
    'bootstrap': [True],
}

rf = RandomForestRegressor(random_state=42)

# Step 2: Hyperparameter tuning (GridSearchCV)
grid_search = GridSearchCV(rf, param_grid, cv=10, scoring='r2', n_jobs=6)
grid_search.fit(X_train, y_train)

# Get the best hyperparameters
best_rf = grid_search.best_estimator_

# Step 3: Feature selection using RFECV
rfecv = RFECV(estimator=best_rf, step=1, cv=10, scoring='r2')
rfecv.fit(X_train, y_train)

# Get the selected features
selected_features = X_train.columns[rfecv.support_]

# Subset your data to include only the selected features
X_train_selected = X_train[selected_features]
X_test_selected = X_test[selected_features]

# Step 4: Fit the Random Forest Regression Model
final_rf = RandomForestRegressor(random_state=42, **grid_search.best_params_) # ** =  dictionary unpacking, it is using the best_params_ attribute of the grid_search object
final_rf.fit(X_train_selected, y_train)

# Step 5: Evaluate the Model (R-squared and RMSE) on Train and Test Data
y_train_pred = final_rf.predict(X_train_selected)
y_test_pred = final_rf.predict(X_test_selected)

r2_train = r2_score(y_train, y_train_pred)
rmse_train = np.sqrt(mean_squared_error(y_train, y_train_pred))
r2_test = r2_score(y_test, y_test_pred)
rmse_test = np.sqrt(mean_squared_error(y_test, y_test_pred))

print(f"Selected Features: {selected_features}")
print("Best Hyperparameters:")
print(grid_search.best_params_)
print(f"R-squared (Train): {r2_train:.4f}")
print(f"RMSE (Train): {rmse_train:.4f}")
print(f"R-squared (Test): {r2_test:.4f}")
print(f"RMSE (Test): {rmse_test:.4f}")
````

Calculating [permutation_importance](https://scikit-learn.org/stable/modules/generated/sklearn.inspection.permutation_importance.html): the permutation importance of a feature is calculated as follows. First, a baseline metric, defined by scoring, is evaluated on a (potentially different) dataset defined by the X. Next, a feature column from the validation set is permuted and the metric is evaluated again. The permutation importance is defined to be the difference between the baseline metric and metric from permutating the feature column.

````python
# Calculate permutation importances
result = permutation_importance(final_rf, X_train_selected, y_train, n_repeats=30, random_state=0)

importances = result.importances_mean
std = result.importances_std
indices = np.argsort(importances)[::-1]

# Explicitly create a figure and axes
fig, ax = plt.subplots(figsize=(10, 6))

# Create the bar plot on the axes
ax.bar(range(X_train_selected.shape[1]), importances[indices], color="b", yerr=std[indices], align="center", alpha=0.5)
ax.set_xticks(range(X_train_selected.shape[1]))
ax.set_xticklabels(X_train_selected.columns[indices], rotation=45, ha='right', size = 16)  # Rotate labels at a 45-degree angle
ax.set_xlabel("")
ax.set_ylabel("Importance")
ax.set_title("Global Feature Importance (Permutation)", size = 18)

# Save the figure
fig.savefig("permutation_importance.png", bbox_inches='tight', dpi=300)

# Show the plot
plt.show()
````

Creating the [shap.Explainer](https://shap.readthedocs.io/en/latest/generated/shap.Explainer.html): uses Shapley values to explain any machine learning model or python function.

````python
# Run SHAP analysis 
X_selected_df = pd.concat([X_train_selected, X_test_selected], axis=0)
explainer = shap.Explainer(final_rf)
shap_values = explainer(X_selected_df)
# dump(shap_values, 'shap_values.joblib')
````

Creating a [shap.plots.bar](https://shap.readthedocs.io/en/latest/example_notebooks/api_examples/plots/bar.html): passing a matrix of SHAP values to the bar plot function creates a global feature importance plot, where the global importance of each feature is taken to be the mean absolute value for that feature over all the given samples.

````python
# Plot SHAP bar
shap.plots.bar(shap_values, show=False) # Set show=False to customize the plot before showing it

# Get the current axes
ax = plt.gca()

# Set transparency for the bars
for patch in ax.patches:
    patch.set_alpha(0.7)

# Customizing the plot using matplotlib functions
plt.title("Global Feature Importance (SHAP)", size = 18)  # Adding a title
plt.xlabel('SHAP Value (impact on RBR)')  # Setting the X-axis label
plt.ylabel('')  # Setting the Y-axis label

# Tried to adjust the figure size or transparency here but didn't work 

# Save the figure
plt.savefig("Global_importance_shap.png", bbox_inches='tight', dpi=300)

# Finally, show the plot
plt.show()
````

Creating a [shap.plots.beeswarm](https://shap.readthedocs.io/en/latest/example_notebooks/api_examples/plots/beeswarm.html): the beeswarm plot is designed to display an information-dense summary of how the top features in a dataset impact the model’s output. Each instance of the given explanation is represented by a single dot on each feature row. The x position of the dot is determined by the SHAP value (shap_values.value[instance,feature]) of that feature, and dots “pile up” along each feature row to show density. Color is used to display the original value of a feature (shap_values.data[instance,feature]). In the plot below we can see that Rx (prescribed fire) is the most important feature on average, and than absence (blue) of Rx is less likely to decrease burn severity (RBR).

````python
# Plot SHAP beeswarm
shap.plots.beeswarm(shap_values, show = False)  # Set show=False to customize the plot before showing it

# Customizing the plot using matplotlib functions
plt.title("Local Feature Importance (SHAP)", size = 18)  # Adding a title
plt.xlabel('SHAP Value (impact on RBR)')  # Setting the X-axis label
plt.ylabel('')  # Setting the Y-axis label


# Save the figure
plt.savefig("Local_importance_shap.png", bbox_inches='tight', dpi=300)

# Finally, show the plot
plt.show()
````

The local feature importance quantifies the contribution of each feature to the prediction for an individual observation. These contributions can be positive or negative, which indicates whether the effect of the feature increases or decreases the predicted outcome, respectively. In the context of burn severity (RBR), a negative SHAP value for a feature would suggest that the value of that feature is associated with a decrease in burn severity, while a positive SHAP value would suggest an increase. The magnitude of the SHAP value reflects the strength of the feature's impact on the prediction, with larger absolute values indicating a greater predicted influence.

Creating a [shap.plots.scatter](https://shap.readthedocs.io/en/latest/generated/shap.plots.scatter.html): plots the value of the feature on the x-axis and the SHAP value of the same feature on the y-axis. This shows how the model depends on the given feature, and is like a richer extension of classical partial dependence plots. Vertical dispersion of the data points represents interaction effects. The scatter plot points are colored by the feature that seems to have the strongest interaction effect with the feature given by the shap_values argument.

The interaction effect between two variables represents how their combination affects the prediction of burn severity, which cannot be fully explained by looking at each feature individually

In this case we are interested in further exploring prescribed fire (Rx) and discover with which feature it had the strongest interaction.

````python
# Plot SHAP interaction for Rx
shap.plots.scatter(shap_values[:, "Rx"], color=shap_values, show = False, alpha=0.6)  # Set show=False to customize the plot before showing it

# Customizing the plot using matplotlib functions
plt.title("Rx' Local Feature Interaction", size = 18)  # Adding a title
plt.xlabel('Prescribed fire (Rx)')  # Setting the X-axis label
plt.ylabel('SHAP Value (impact on RBR)')  # Setting the Y-axis label

# Tried to adjust the figure size or transparency here but didn't work 
# For example:
# fig = plt.gcf()
# fig.set_size_inches(10, 8)


# Save the figure
plt.savefig("Local_interaction_Rx_shap.png", bbox_inches='tight', dpi=300)
# Finally, show the plot
plt.show()
````

In this SHAP interaction plot, each point represents an individual prediction from the model, illustrating the interaction between percent canopy cover (pct_canopy_cover) and prescribed fire (Rx) on the predicted outcome, which is the burn severity (RBR). The x- and y-axis show true values of percent canopy cover and SHAP values, respectively. Points above zero suggest that higher percent canopy cover increases predicted burn severity, whereas points below zero suggest a decrease.

The color indicates the value of prescribed fire for each observation, with a scale ranging from blue to red. While the prescribed fire's scale includes 0.5, the only possible values are 0 (absence) and 1 (presence). For example, a blue point with a higher positive SHAP value suggests that when Rx is absent, percent canopy cover substantially increases the predicted burn severity. Conversely, a red point with a lower positive SHAP value implies that when prescribed fire is present, percent canopy cover has a lower impact on predicted burn severity.

## LESSONS LEARNED
**Model's performance**
Model tuning does affects model perfomance and it's worth it to explore multiple hyperparameters options. This however can be computationally expensive. 

The model resulted into a higher R-squared and lower RMSE on the training data compared to the test data, which is a common occurrence as models tend to perform better on data they were trained on. The moderate decrease in R-squared and increase in RMSE for the test data suggest a degree of overfitting, indicating the model may not generalize as well to new data. However, given the model's specific purpose — to find key drivers of burn severity for a well-defined geographic area — the moderate overfitting observed is not a major concern. The focus on varibale importance for this particular context allows for some flexibility in model fit. Nevertheless, should the model be used for predictive purposes, continued monitoring of its predictive power on new data would be essential.

SHAP analysis offered great insights on how the black-box model performed and should be interpreted, by offering global and local perspectives. 

**Key drivers of burn severity and the role of prescribed fire**

Key drivers of burn severity included prescribed fire, coarse-scale topographic position index, evaporative stress index, max temperature by fire progression day, and percent canopy cover. While the entire Bootleg Fire was driven by extreme fire weather (a top-down control), analyzing burn severity over a relatively small study area allowed us to uncover the importance of bottom-up controls such as prescribed fire, topography, and forest structure.
 
Fuel treatments had a significant impact on burn severity, and the use of prescribed fire was effective in mitigating burn severity. The pronounced influence of this variable, especially in its interaction with canopy cover, suggests that prescribed fire may have served as an indicator for the presence of ground and surface fuels. Where this treatment was conducted, the diminished impact (reflected by smaller, positive SHAP values) of higher canopy cover suggests that the fire was less likely to reach the tree canopies, and that therefore there was less fuel avaible in the lower strata. Even with 70% canopy cover, prescribed fire treatments led to reduced burn severity. Conversely, in the absence of prescribed fire, higher canopy cover significantly increased burn severity, as evidenced by larger, positive SHAP values. Meaning that, potentially, lower fuel strata served as ladder fuels, allowing the fire to reach the upper canopy stratum and burn tree crowns. While ground validation may be necessary to draw definitive conslusions, these results suggest that burn severity was driven by surface fuels.

While all categories of predictors have some importance in explaining burn-severity variability, only fuel treatments and forest structure are under our control, to some extent. Thus, understanding which variables within these two categories are the most important will serve as a valuable resource for land managers seeking to establish fire regimes that promote forest resilience through strategic fuel load reduction.
